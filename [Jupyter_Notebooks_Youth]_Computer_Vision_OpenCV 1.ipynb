#### Image Smoothing
- Averaging
- Guasian Blurring
- Median Blurring
- Bilateral Filtering

```python
# Averaging
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('Images/OpenCV-open-source-new-version-released.jpeg')
blur = cv.blur(img,(15,15))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred/Smoothened')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Gaussian Blurring
img = cv.imread('Images/dog-collie-beach.jpeg')
blur = cv.GaussianBlur(img,(15,15),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred/Smoothened')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Median Blurring
img = cv.imread('Images/Noisy Image.png')
blur= cv.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred/Smoothened')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Median Blurring
img = cv.imread('Images/Noisy Image-2.png')
blur = cv.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred/Smoothened')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Resizing Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#Resizing a image
image = cv.imread("Images/OpenCV-open-source-new-version-released.jpeg", 1)
# Loading the image

half = cv.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv.resize(image, (1050, 1610))

stretch_near = cv.resize(image, (780, 540),
               interpolation = cv2.INTER_NEAREST)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```

## Tresholding

```python
#https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html
#Simple Tresholding
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('Images/dog-collie-beach.jpeg',0)
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#Adaptive Tresholding
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('Images/dog-collie-beach.jpeg',0)
img = cv.medianBlur(img,5)
ret,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,2)
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## Image Egde Detection

#### Sobel Filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# loading image
#img0 = cv2.imread('SanFrancisco.jpg',)
img0 = cv2.imread('Images/Valve_original_(1).png',)

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

#### Laplacian

```python
# loading image
#img0 = cv2.imread('SanFrancisco.jpg',)
img0 = cv2.imread('Images/Valve_original_(1).png',)

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels
laplacian = cv2.Laplacian(img,cv2.CV_64F)


plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.show()
```

#### Canny Edge Detection

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('Images/Canny_Detector_Tutorial_Original_Image.jpeg',0)
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

Write a Python program to apply the following image tranformations to **Image/bigben.jpg** image:
<br>**1. Apply Canny edge detection**
<br>**2. Resize the image**
<br>View the image.

```python

import cv2 as cv
import matplotlib.pyplot as plt

original_img = cv.imread('Images/bigben.jpg')

# Resize
resized = cv.resize(original_img, (0,0), fx=0.5, fy=0.5)

# Canny edge detection
gray = cv.cvtColor(resized, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray, 100, 200)

# View
plt.subplot(131), plt.imshow(cv.cvtColor(original_img, cv.COLOR_BGR2RGB))
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(cv.cvtColor(resized, cv.COLOR_BGR2RGB))
plt.title('Resized'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(edges, cmap='gray')
plt.title('Canny Edge'), plt.xticks([]), plt.yticks([])

plt.show()
```

Write a Python program to apply the following image tranformations to **Image/bigben.jpg** image:
<br>**1. Read the image**
<br>**2. Apply gaussian blur**
<br>**3. Apply Laplacian image edge detection to the image with proper preprocessing**

```python

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('Images/bigben.jpg')

# Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Gaussian blur
blur = cv.GaussianBlur(gray, (3,3), 0)

# Laplacian
laplacian = cv.Laplacian(blur, cv.CV_64F)

# View
plt.subplot(131), plt.imshow(gray, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(blur, cmap='gray')
plt.title('Gaussian Blur'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.show()
```
