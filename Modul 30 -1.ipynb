### String Datatype

```python
# Create a string variable 'name' and set it to value 'Bob'
name = 'Bob'
```

```python
# print the value inside the variable 'name'
print(name)
```

```python
# check the datatype of variable 'name'
type(name)
```

```python
#String Concatenation

a = "Learning"
b = "Python"

#Combine string a and b

c = a + b # '+' operator can be used to concatenate two strings
print(c)
```

```python
#String Concatenation

a = "Learning"
b = "Python"

#Combine string a and b: '+' operator can be used to concatenate two strings

c = a +" "+ b
print(c)
```

```python
# String Slicing

a = "Welcome to Python Programming"

#Get the word Python - character indexing starts from 0

print(a[11:17])
```

```python
# String Formatting

slices = 2
item = 'cake'
price = 50

text = "I need {} slices of {} for {} dollars!"
order = text.format(slices,item,price)
print(order)
```

```python
# Ask user for name

name = input("What is your name?: ")
umur = input("berapa umur kamu?")
kota = input("dimana kamu tinggal?")
minat = input("apa yang kamu sukai")

# Similarly, Ask user for age


#Ask user for city


#Ask user what they enjoy


#Create output text using string formatting

string = "Hi {}, you are {} years old. You live in {} and you love {}!"

output = string.format(name,umur,kota,minat)

#Print output to screen
print(f"${output}")
```
