### Exponent functions using SciPy

#### exponential of the given value

```python
# import scipy special which includes functions such as exponential,trigonometric etc.
from scipy import special
# import warnings to avoid any version warnings etc.
import warnings
warnings.filterwarnings("ignore")
```

```python
# find an exponential value of 2^5 with SciPy
exponent=5
Expvalue=special.exp2(exponent)
print(Expvalue)
```

```python
# find exponential value using Scipy
Expresult = special.exp10([2, 9])
print(Expresult)
```

#### relative error of the Exponential function

```python
# find the relative error of the exponential function
Relerror = special.exprel([-0.25, -0.1, 0, 0.1, 0.25])
print(Relerror)
```

#### Log sum exponential function

```python
# find the log sum exponential values
import numpy as np
a = np.arange(10)
res = special.logsumexp(a)
print(res)
```

### Statistics functions using SciPy

```python
# import stats functions
from scipy import stats
```

#### Descriptive Statistics

##### gmean()

```python
# this calculates the geomteric mean of the given data
data=[4,5,2,9,8,10]
print(stats.gmean(data))
```

##### hmean()

```python
# this calculates the harmonic mean of the given data
data = [3,4,5]
print(stats.hmean(data))
```

##### kurtosis()

```python
# Kurtosis explains the occurrence of outliers
data=[4,7,10]
print(stats.kurtosis(data))
```

##### skew()

```python
# Skewness explains symmetric distribution
data=[1,2,90,4,50]
print(stats.skew(data))
```

##### iqr()

```python
# inter quartile region
data=[1,2,90,4,50]
print(stats.iqr(data))
```

##### describe()

```python
# describe()- displays the summary statistics
data=[2,3,4,5,20,90]
print(stats.describe(data))
```

The Forest Cover Type dataset is a well-known dataset provided by scikit-learn (sklearn) for educational and research purposes. It involves the classification of forest cover types based on various attributes.



#### Dataset Description:
The dataset consists of 30x30 meter patches of forest in the United States. The goal is to predict the dominant tree species in each patch, resulting in a multiclass classification problem with seven possible cover types. Each patch is represented by 54 features, a mix of boolean indicators, discrete, and continuous measurements.


#### Information about the Dataset:
The dataset involves predicting forest cover type based solely on cartographic variables, without using remotely sensed data. The actual forest cover type was determined from US Forest Service data, and independent variables were derived from data originally obtained from US Geological Survey and US Forest Service data. The dataset contains both binary (0 or 1) columns for qualitative variables like wilderness areas and soil types.

The study area comprises four wilderness areas in the Roosevelt National Forest of northern Colorado, each representing different forest cover types resulting from ecological processes rather than forest management. The primary tree species in these areas are detailed.

#### Attribute Information:
Here are some attributes in the dataset:

- Elevation: Elevation in meters (quantitative).
- Aspect: Aspect in degrees azimuth (quantitative).
- Slope: Slope in degrees (quantitative).
- Horizontal_Distance_To_Hydrology: Horizontal distance to nearest surface water features in meters (quantitative).
- Vertical_Distance_To_Hydrology: Vertical distance to nearest surface water features in meters (quantitative).
- Hillshade_9am, Hillshade_Noon, Hillshade_3pm: Hillshade indices at different times (quantitative).
- Horizontal_Distance_To_Fire_Points: Horizontal distance to nearest wildfire ignition points in meters (quantitative).
- Wilderness_Area (4 binary columns): Wilderness area designation (qualitative).
- Soil_Type (40 binary columns): Soil Type designation (qualitative).
- Cover_Type (7 types): Forest Cover Type designation (integer).
- The dataset's goal is to predict the 'Cover_Type' attribute, which represents the forest cover type with integer values from 1 to 7.

```python
# Q9. Split the digit dataset using scikit-learn
```

```python
X, y = datasets.load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Modeling

```python
import sklearn as sl
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


# Initialize models
lr = LogisticRegression()                   # Logistic Regression model
knn = KNeighborsClassifier()                # K-Nearest Neighbors model
dTree = DecisionTreeClassifier(random_state=12)   # Decision Tree model
clf = RandomForestClassifier(n_estimators=100)     # Random Forest model

# Train and predict using each model
models = [lr, knn, dTree, clf]
model_names = ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest']
best_accuracy = 0
best_model = None

# Iterate through each model and evaluate its performance
for model, name in zip(models, model_names):
    model.fit(X_train, y_train)                # Train the model using the training set
    y_pred = model.predict(X_test)             # Predict using the test set
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
    print(f"{name} Accuracy: {accuracy}")      # Print accuracy for the current model

    # Update the best model if the current model's accuracy is higher
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = name

# Print the best model
print("\nBest Model:", best_model_name)
```
