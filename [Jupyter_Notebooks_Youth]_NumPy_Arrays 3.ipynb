### attribute-astype

```python
# astype- changes the data type from one type to another data type
#In order to change the dtype of the given array object, we will use numpy.astype() function.
#The function takes an argument which is the target data type.
#The function supports all the generic types and built-in types of data.
```

-- Example: Given a numpy array whose underlying data is of 'int32' type. Change the dtype of the given object to 'float64'.

```python
# Create a numpy array
Astypearr1D = np.array([10, 20, 30, 40, 50])
# Print the array
print(Astypearr1D)
print(Astypearr1D.dtype)
```

```python
# change the dtype to 'float64'
Astypearr1D= Astypearr1D.astype('float64')
print(Astypearr1D)
print(Astypearr1D.dtype)
```
