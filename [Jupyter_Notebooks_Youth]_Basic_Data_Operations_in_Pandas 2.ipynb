In this notebook, we are going to discuss about:
- Descriptive Statistical Operations in Pandas: mean, median, mode, count, std, max, min, describe

## Reading a file

```python
# Descriptive Statistical Operations on Data Frame
# Import Pandas Library
import pandas as pd

# Load Titanic Dataset as Dataframe
dataset = pd.read_csv('Datasets/train.csv')

# Show dataset
# head() bydefault show
# 5 rows of the dataframe
dataset.head()
```

## Descriptive Statistical Operations

```python
# dataset no. of columns and rows
dataset.shape
```

### mean

```python
# Calculate mean
mean = dataset['Age'].mean()

# Print mean
print(mean)
```

### median

```python
# Calculate Median of 'Fare' column
median = dataset['Fare'].median()

# Print median
print(median)
```

### mode

```python
# Calculate Mode of 'Sex' column
mode = dataset['Sex'].mode()

# Print mode
print(mode)
```

### Count

```python
# Calculate Count of 'Ticket' column
count = dataset['Ticket'].count()

# Print count
print(count)
```

### Standard Deviation

```python
# Calculate Standard Deviation
# of 'Fare' column
std = dataset['Fare'].std()

# Print standard deviation
print(std)
```

### maximum value

```python
# Calculate Maximum value in 'Age' column
maxValue = dataset['Age'].max()

# Print maxValue
print(maxValue)
```

### minimum value

```python
# Calculate Minimum value in 'Fare' column
minValue = dataset['Fare'].min()

# Print minValue
print(minValue)
```

### describe statistical summary

```python
# Statistical summary
dataset.describe()
```
