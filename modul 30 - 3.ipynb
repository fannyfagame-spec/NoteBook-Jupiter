```python
#Curious Baby Simulator Program using while loop
# Counting the number of vowels and consonants in a particular string using for and if else

text = input("Masukkan sebuah kalimat: ").lower()

vowels = 0
consonants = 0

for char in text:
    if char.isalpha():  # hanya huruf yang dihitung
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Jumlah huruf vokal:", vowels)
print("Jumlah huruf konsonan:", consonants)
```

```python
# Variable Scope in Functions

x = 98 # global variable scope

def f1():
    print(x)

def f2():
    print(x)

f1()
f2()
```

```python
# Variable Scope in Functions

x = 98 # global variable scope

def f1():
    x = 15 # local variable
    print(x)

def f2():
    x = 30 # local variable
    print(x)

f1()
f2()
print(x)
```

```python
# Variable Scope in Functions

x = 98 # global variable scope

def f1():
    global x
    x = 50
    print(x)

def f2():
    x = 30 # local variable
    print(x)

f1()
f2()
print(x)
```

```python
#Creating a function without default parameters

def intro(name,city,interest):
    intro = 'Hi I am {},I live in {} and I love {}'.format(name,city,interest)
    return intro

print(intro("Alice","Mumbai","Music"))

print(intro("Ryan","Delhi")) # missing argument will give an error
```

### Simple Calculator Program

* There will be 4 available options in the calculator program:
    * Addition
    * Subtraction
    * Multiplication
    * Division
* Take two numbers as input from the user, apply if… elif… else conditional statements for performing the required operation
* Use add(), subtract(), multiply() and divide() function for evaluation the respective operation in the calculator.

Reference from the content slide- Module 30


```python

# A Simple Calculator Program

# Create a function for adding two numbers
def add(a, b):
    return a + b

# Create a function for subtracting two numbers
def subtract(a, b):
    return a - b

# Create a function for multiplying two numbers
def multiply(a, b):
    return a * b

# Create a function for dividing two numbers
def divide(a, b):
    return a / b

# Display the list of operations using print
print("Please select the operation.")
print("a. Add")
print("b. Subtract")
print("c. Multiply")
print("d. Divide")

# Take the input from the user
op = input("Please enter choice (a/ b/ c/ d): ")

# Take 2 numbers as input from the user and convert into integer format
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Logic for performing the specified operation
if op == "a":
    print("Result:", add(num1, num2))

elif op == "b":
    print("Result:", subtract(num1, num2))

elif op == "c":
    print("Result:", multiply(num1, num2))

elif op == "d":
    if num2 != 0:
        print("Result:", divide(num1, num2))
    else:
        print("Error: Cannot divide by zero.")

else:
    print("Invalid operation.")
```
