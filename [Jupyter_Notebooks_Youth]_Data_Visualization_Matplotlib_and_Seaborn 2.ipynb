In this notebook, we are going to discuss how to visuzlise data using matplotlib and seaborn to better our understanding of the dataset before we use it in modelling.

```python
# Import necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# ignore any warnings:, such as updates etc..
import warnings
warnings.filterwarnings('ignore')
```

```python
# Simple bar chart
# Create dataset
height = [3, 12, 5, 18, 45]
bars = ['A', 'B', 'C', 'D', 'E']
plt.bar(bars,height,color='aquamarine')
plt.xlabel('bars')
plt.ylabel('height')
plt.show()
```

### Matplotlib

#### Bar Plots

##### Vertical Bar Plots

```python
# Dataset generation: # How many students like each of these bakery items: Pastries, Chocolates, Doughnuts, Pizza
students_dict = {'Pastries':33, 'Chocolates':28, 'Doughnuts':30, 'Pizza':20}
Bakeryitems = list(students_dict.keys())
values = list(students_dict.values())
# bar plot properties
fig = plt.figure(figsize = (4, 4))
plt.bar(Bakeryitems, values,color='green',align='center')
plt.xlabel("Bakery items")
plt.ylabel("No. of students who like the items")
plt.title("Students survey for bakery items")
plt.show()
```

##### Horizontal Bar Plots

```python
# Dataset generation:# How many students like each of these bakery items: Pastries, Chocolates, Doughnuts, Pizza
students_dict = {'Pastries':33, 'Chocolates':28, 'Doughnuts':30, 'Pizza':20}
Bakeryitems = list(students_dict.keys())
values = list(students_dict.values())
fig = plt.figure(figsize = (3,3))
plt.barh(Bakeryitems, values, color ='green',align='center')
plt.xlabel("No. of students who like the items")
plt.ylabel("Bakery items")
plt.title("Students survey for bakery items")
plt.show()
```

##### Grouped Vertical Bar Plots

```python
# Dataframe creation: Classwise survey about the interest of the students for the bakery items
Class_survey=pd.DataFrame({"Class 1":[25,20,35,30],"Class 2":[15,10,37,40],"Class 3":[25,12,30,25]},
                        index=["Pastries","Pizza","Doughnuts","Chocolates"])
print(Class_survey)# print the DataFrame.
#  Bar plot properties
Class_survey.plot(kind="bar",figsize=(4, 4),align='center')
plt.title("Class-wise Survey")
plt.xlabel("Bakery items")
plt.ylabel("Class")
plt.show()
```

##### Grouped Horizontal Bar Plots

##### Stacked Vertical Bar Plots

```python
# Dataframe creation: Classwise survey about the interest of the students for the bakery items
Class_survey=pd.DataFrame({"Class 1":[25,20,35,30],"Class 2":[15,10,37,40],"Class 3":[25,12,30,25]},
                        index=["Pastries","Pizza","Doughnuts","Chocolates"])
print(Class_survey)# print the DataFrame.
#  Bar plot
Class_survey.plot(kind="bar",stacked ='True',figsize=(6,6),align='center')
plt.title("Class Survey")
plt.xlabel("Bakery items")
plt.ylabel("Class")
plt.show()
```

```python
# Scatter plot with different shape and colour for two datasets.
# dataset-1: Number of fruits and vegetables liked by students.
fruits = [5, 8, 6, 9, 3, 7, 2, 4]
vegetables = [10, 6,8,7,10,4,9,3]
# dataset-2: Number of pencils and pens with the students.
pencils = [3,7,12,20,10,7,5,9]
pens = [5,8,4,9,6,7,3,10]
plt.scatter(fruits, vegetables, c ="green",linewidths = 2,marker ="o",edgecolor ="black",s = 50)
plt.scatter(pencils, pens, c ="yellow",linewidths = 2,marker ="^",edgecolor ="black",s = 200)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Box Plots

```python
# Creating dataset: Generate a box plot with 1 to 100 range
values=[4,3,5,7,10,45,20,40,90,70]
fig = plt.figure(figsize =(5,5))
# Creating plot
plt.boxplot(values)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
# show plot
plt.show()
```

```python
# Creating dataset: Generate a box plot for different subjects for students in a class
English=[45,40,55,75,80,85,90,70]
Maths=[55,80,65,85,60,75,90,50]
Science=[50,40,45,65,70,85,60,55]
values=[English, Maths, Science]
# Creating plot
fig = plt.figure(figsize =(5,5))
box=plt.boxplot(values,patch_artist=True, labels=['English','Maths','Science'])
# applying color to the individual subjects in the boxes
colors = ['green', 'maroon', 'gray']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject-wise Marks")
# show plot
plt.show()
```

```python
# Dataset creation: What is the mode of payment by the customers in a restaurant day wise?
# C- represents cash
# O- represents online mode
Customer_survey=pd.DataFrame({'Day': ['Mon', 'Tue', 'Wed', 'Thur', 'Fri','Mon', 'Tue', 'Wed', 'Thur', 'Fri'],
                   'Customers': [44, 46, 49, 59, 44, 34, 26, 59, 60,54],
                   'Payment mode':['C','O','C','O','O', 'O','C','O','C','C']})
print(Customer_survey)
#create grouped bar chart
sns.barplot(x='Day', y='Customers', hue='Payment mode',data=Customer_survey,palette=['purple', 'green'])
plt.xlabel("Day-wise")
plt.ylabel("Number of Customers")
plt.title("Payment mode of Customer in an restaurant")
plt.show()
```

```python
#Q2. Write a Python programming to display a scatter plot using Matplotlib with the following Sample data:
#Dataset1: Number of tickets and amount paid for attending movie
#Numberoftickets = [10, 5, 9, 15, 20, 8]
#Amountpaid=[200,100,180,300,400,160]
#Dataset2: Number of pizzas and fries ordered in the movie theatre
#Pizza=[25,15, 20, 12, 10,30]
#Fries=[15,10,12,25,20,35]
```

```python
Numberoftickets = [10, 5, 9, 15, 20, 8]
Amountpaid = [200, 100, 180, 300, 400, 160]
Pizza = [25, 15, 20, 12, 10, 30]
Fries = [15, 10, 12, 25, 20, 35]

plt.scatter(Numberoftickets, Amountpaid, c='blue', marker='o', edgecolor='black', label='Tickets vs Amount')
plt.scatter(Pizza, Fries, c='red', marker='^', edgecolor='black', label='Pizza vs Fries')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
#Q3. Write a Python programming to display a box plot using Seaborn with the following Sample data:
#Annie=[12,7,3,9,20,15,25]
#Bob=[15,8,25,5,20,6,10]
#Caty=[25,14,24,26,30,15,30]
```

```python
Annie = [12, 7, 3, 9, 20, 15, 25]
Bob = [15, 8, 25, 5, 20, 6, 10]
Caty = [25, 14, 24, 26, 30, 15, 30]

values = [Annie, Bob, Caty]
fig = plt.figure(figsize=(5, 5))
sns.boxplot(data=values)
plt.xticks([0, 1, 2], ['Annie', 'Bob', 'Caty'])
plt.xlabel("Name")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

```python
#Q4. Write a Python programming to display a violin plot using Seaborn with the following Sample data:
#Annie=[12,7,3,9,20,15,25]
#Bob=[15,8,25,5,20,6,10]
#Caty=[25,14,24,26,30,15,30]
```

```python
Annie = [12, 7, 3, 9, 20, 15, 25]
Bob = [15, 8, 25, 5, 20, 6, 10]
Caty = [25, 14, 24, 26, 30, 15, 30]

values = [Annie, Bob, Caty]
fig = plt.figure(figsize=(5, 5))
sns.violinplot(data=values)
plt.xticks([0, 1, 2], ['Annie', 'Bob', 'Caty'])
plt.xlabel("Name")
plt.ylabel("Values")
plt.title("Violin Plot")
plt.show()
```
