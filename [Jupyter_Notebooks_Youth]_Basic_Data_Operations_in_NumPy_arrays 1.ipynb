## Section 1: Arithmetic operations in NumPy
### NumPy add function and operator

```python
# import numpy library
import numpy as np
```

```python
# Create two arrays of same size and dimensions
# NumPy Add function
AddarrA=np.array([10,20,30,40])
AddarrB=np.array([1,2,3,4])
np.add(AddarrA,AddarrB)
```

```python
# NumPy add operator
AddarrA,AddarrB
```

```python
# add arrays using '+' operator
AddarrA+AddarrB
```

### NumPy subtract function and operator

```python
# Create two arrays of same size and structure
# NumPy Subtract function
SubtractarrA=np.array([25,15,35,45])
SubtractarrB=np.array([10,6,11,18])
np.subtract(SubtractarrA,SubtractarrB)
```

```python
# NumPy subtract operator
SubtractarrA,SubtractarrB
```

```python
# subtract  arrays using '-' operator
SubtractarrA-SubtractarrB
```

### NumPy multiply function and operator

```python
# Create two arrays of same size and structure
# NumPy Mutiply function
MultiplyarrA=np.array([3,5,7,9,11])
MultiplyarrB=np.array([2,4,6,8,10])
np.multiply(MultiplyarrA,MultiplyarrB)
```

```python
# NumPy multiply operator
MultiplyarrA,MultiplyarrB
```

```python
# multiply two arrays using '*' operator
MultiplyarrA*MultiplyarrB
```

### NumPy divide function and operator

```python
# Create two arrays of same size and structure
# NumPy Divide function
DividearrA=np.array([10,15,27,39,51])
DividearrB=np.array([2,3,9,13,16])
np.divide(DividearrA,DividearrB)
```

```python
# NumPy divide operator
DividearrA,DividearrB
```

```python
# divide arrays using '/' operator
DividearrA/DividearrB
```

## Section 2: Matrix operations in NumPy
### Inner Product

```python
# Vectors as 1D numpy arrays
# Create two vectors
InnerarrA = np.array([1, 2, 3])
InnerarrB = np.array([4, 5, 6])
print(InnerarrA,InnerarrB)
```

```python
# Calculate inner product
np.inner(InnerarrA,InnerarrB)
```

```python
# Calculate the dot product
np.dot(InnerarrA,InnerarrB)
```

### Dot Product

```python
# Matrices as ndarray objects
DotarrA = np.array([[1, 2], [3, 4]])
DotarrB = np.array([[5, 6, 7], [8, 9, 10]])
print(DotarrA)
```

```python
print(DotarrB)
```

```python
# apply dot product
np.dot(DotarrA,DotarrB)
```

### Transpose of a matrix

```python
#Create an array
Newarr = np.array([[1, 2], [3, 4], [5, 6]])
Newarr
```

```python
# Apply transpose of a matrix
np.transpose(Newarr)
```

## Section 3: Statistical Operations in NumPy

### mean

```python
# Create an array
ArrMean = np.array([5,6,7])
ArrMean
```

```python
# mean
print(np.mean(ArrMean))
```

### median

```python
# create array
ArrMedian=np.array([5,6,7])
ArrMedian
```

```python
# median
print(np.median(ArrMedian))
```

### Standard Deviation

```python
# create array
ArrStdDev=np.array([5,6,7])
ArrStdDev
```

```python
# standard deviation
print(np.std(ArrStdDev))
```
